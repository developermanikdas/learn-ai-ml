```python
#4 Create Json and update

import json
my_dict = {"Kolkata": 10000, "Hitech_city": 20000, "Mumbai": 100000}

with open("cities2.json", "x+") as f:
    json.dump(my_dict, f)
    f.seek(0)
    loaded_cities = json.load(f)
    print(loaded_cities)

    for city, population in loaded_cities.items():
        print(f"{city}: {population:,}")
        print("-" * 32 + "\n")

new_city = input("Enter the name of a new city: ")
new_population = int(input(f"Enter the population for {new_city}: "))

my_dict[new_city] = new_population

with open("cities.json", "w") as file:
    json.dump(loaded_cities, file, indent=4)

print(f"\nSuccess! {new_city} has been added to 'cities.json'.")
```
