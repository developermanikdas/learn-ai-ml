```python
#1 File operation write names and show those

f = open("names.txt", "w")
f.write("Manik \nDevi Ji \nJohn \nAnirban \nHappy")
f = open("names.txt", "r")
data = f.read()
print(data)
f.close()
```

```python
### Practice problem

with open("poem.txt", "r") as f:
    data = True
    word = "crimson"
    line = 1
    
    while data:
        data = f.readline()
        if word in data:
            print(line)
        line += 1   
```
