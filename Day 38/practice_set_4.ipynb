```python
#2 Create a class called book with title author and a list of review parameters. Add the methods like add review Sound review show review. 

class Book:
    def __init__(self, title, author, list_of_review):
        self.title = title
        self.author = author
        self.list_of_review = list_of_review

    def add_review(self, review):
        self.list_of_review.append(review)
        return "Review Added"

    def count_review(self):
        return len(self.list_of_review)

    def show_review(self):
        return self.list_of_review    
        
b1 = Book("The subtle art of not giving a F*ck", "Mark Manson", ["Great", "Super", "Wow"])   

print(b1.add_review("True love"))
print(b1.show_review())
print(b1.count_review())
```
