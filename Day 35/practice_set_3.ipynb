```python
#1 WAF to check palindrome 

my_string = input("Enter a string: ")
clean_text = my_string .replace(" ", "").lower()

def is_palindrome(my_string):
    
#     return clean_text == clean_text[::-1]

    start = 0
    end = len(clean_text) - 1
    
    while start < end:
        if clean_text[start] == clean_text[end]:
            start+=1
            end-=1
        else:    
            return False
    return True
    
is_palindrome(my_string)
```

```python
#2 Sum of all nums in a list

my_list = [1,3,3,56,7,87]

def my_avg(numbers):
    return (sum(numbers) / len(numbers))

print(my_avg(my_list))
```

```python
#3 Input to lists and merge them into one. Then after that sort them 
import random

list1 = []
list2 = []

while len(list1) < 3:
   list1.append(random.randint(1,100))
    
while len(list2) < 3:
   list2.append(random.randint(1,100))    

total = sorted(list1 + list2)
total
```

```python
#4 Separate even an odd number from a temple. 

my_tup = (1, 2, 3, 5, 6, 7, 8)

even_tuple = tuple(x for x in my_tup if x%2 == 0)
odd_tuple = tuple(x for x in my_tup if x%2 != 0)

print(*even_tuple)
print(*odd_tuple)
```
