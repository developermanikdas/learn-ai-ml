```python
import requests
from bs4 import BeautifulSoup

import pandas as pd

page_count = 1

data_list = []

while True:
    URL = f"https://quotes.toscrape.com/page/{page_count}"
    res = requests.get(URL)
    soup = BeautifulSoup(res.text, "lxml")
    quotes = soup.select("div.quote")
    
    if not quotes:
        print(f"No Page left in page {page_count}")
        break
    else:
        for quote in quotes:
            quote_text = quote.find("span", class_="text").text
            quote_author = quote.find("small", class_="author").text
            quote_tags = quote.find_all("a", class_="tag")
            for quote_tag in quote_tags:                
                if quote_tag.text == "life":
                    # print(quote_text, quote_author)
                    # print("\n")
                    data_list.append({"quote": quote_text, "author": quote_author})
    page_count += 1


df = pd.DataFrame(data_list)
df.to_csv("quotes.csv", index=False)

df = pd.read_csv('quotes.csv')
df = 
```
