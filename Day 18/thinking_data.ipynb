```python
import json

def load_data(filename, mode):
    with open(filename, mode) as file:
        data = json.load(file)
        return data

#cleaning and structuring data

raw_data = load_data("data.json", "r")

raw_numbers = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5}

def clean_rating(data):
    unique_user = []
    unique_name = set()
    
    for obj in data:
        raw_rating = str(obj["rating"]).strip().lower()
        
        if raw_rating in raw_numbers:
            raw_rating = raw_numbers[raw_rating]
        obj["rating"] = raw_rating

        raw_age = obj.get("age")
        if (raw_age == None):
            obj["age"] = None

        if(obj["name"] not in unique_name):
            unique_name.add(obj["name"]) 
            unique_user.append(obj)

    return unique_user  

cleaned_data = clean_rating(raw_data)

def cal_avg_rating(data):
    total = 0
    low_rating = 0
    for user in data:
        total = total + float(user["rating"])
        if(float(user["rating"]) < 3):
            low_rating += 1
            
    print(total/len(data))    
    print(low_rating/len(data)*100)    

cal_avg_rating(cleaned_data)    
        
        
```

```python
def recommendation(data):
    recommendations = []

    for user in data:
        curr_user = {}
        curr_user["name"] = user["name"]
        if(float(user["rating"]) >= 4):
            curr_user["brand"] = "Apple"
        else:
            curr_user["brand"] = "Samsung"
        recommendations.append(curr_user)
    return recommendations    
recommendation(cleaned_data)    
            
```
