```python
### Practice 
import json

def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
        return data
```

```python
data = load_data("store_data.json")
text_to_num = {"one": 1.0, "two": 2.0, "three": 3.0, "four": 4.0, "five": 5.0}
cleaned_data = []
unique_user = set()

def clean_data(data):

    for user in data:
        
        #1 convert rating to float
        raw_rating = user["rating"].strip().lower()
        if raw_rating in text_to_num:
            user["rating"] = text_to_num[raw_rating]
        else:
            user["rating"] = float(user["rating"])
        

        #2 missing age    
        raw_age = user.get("age")
        if raw_age == None:
            user["age"] = 24  #hardcode avg value, will fix later
        user["age"] = float(user["age"])    


        #3 deduplication 
        if user["name"].strip() in unique_user:
            continue
        unique_user.add(user["name"])    
        cleaned_data.append(user)    
        
clean_data(data)    
```

```python
# get insights 

def get_insights(data):

    total_rating = 0
    poor_rating = 0
    for user in data:

        #1 average rating
        rating = user["rating"]
        total_rating += rating
        avg_rating  = total_rating / len(data)
        
        #2  percentage of poor rating

        if rating < 3:
            poor_rating += 1  
        poor_rating_percent = poor_rating/len(data)*100    

    print(f"Rating average: {avg_rating}")
    print(f"Poor rating percentage: {poor_rating_percent}")
    
get_insights(cleaned_data)        
```

```python
# recomend products

recommendation = [] 

def get_recommendation(data):
    
    for user in data:
        curr_recommendation = {}
        curr_recommendation["name"] = user["name"]
        if user["rating"] >= 4:
            curr_recommendation["brand"] = "Apple"
        else:
            curr_recommendation["brand"] = "Samsung" 
        recommendation.append(curr_recommendation)     

    return recommendation
            
get_recommendation(cleaned_data)            
```
