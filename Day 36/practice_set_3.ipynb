```python
#5 Create a dict and perform operation

studs = {"Manik": 99, "Goddess": 100, "Rahul": 45}

print("Options: (A) Add, (B) Update, (C) Show Specific, (D) Show All")
operation = input("Enter operation name: ").lower().strip()

match operation:
    case "a":
        name = input("Enter student name: ")
        studs[name] = int(input("Enter student mark: "))
        print(f"{name} added")
    case "b":
        name = input("Enter student name: ")
        if name in studs:
            studs[name] = int(input("Enter student updated mark: "))
        else:
            print("No student found.")
    case "c":
        name = input("Enter student name: ")
        if name in studs:
            result = {k: v for k, v in studs.items() if k==name}
            print(result)
        else:
            print("No student found.")    
    case "d":
        print(studs)   
    case _:
        print("Invalid Option")
        
        
        
```

```python
#9 print nums from a list if appear > 1

list1 = [1, 2, 3, 4, 2, 3, 6, 1, 1]

seen = set()
duplicate = set()

for i in list1:
    if i in seen:
        duplicate.add(i)
    else:
        seen.add(i)

print(duplicate)       
```

```python
#10 All unique characters and their account 

my_string = input("Enter a string: ")

unique_chars = set()

for i in my_string:
    unique_chars.add(i)

print(unique_chars)
```
