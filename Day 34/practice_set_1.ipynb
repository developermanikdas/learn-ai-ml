```python
#2 Take two numbers as input from the user and print their sum, difference, product, and quotient

num1 = int(input("Enter Number: "))
num2= int(input("Enter Number: "))

addition = num1+num2
difference = num1-num2
product = num1*num2
quotient = num1/num2

print(addition, difference, product, quotient)
```

```python
#6 swap 2 nums

num1 = int(input("Enter Number: "))
num2= int(input("Enter Number: "))

print(f"Before: Num1 is {num1} and num2 is {num2}")

temp = num1
num1 = num2
num2 = temp

print(f"After: Num1 is {num1} and num2 is {num2}")
```

```python
#7 Ask the user to enter a temperature in Celsius, convert it into Fahrenheit. 

celsius = float(input("Enter temp in celsius: "))
fahrenheit =  celsius*(9/5)+32
print(f"{celsius}\u00B0 Celsius is {fahrenheit} fahrenheit")
```

```python
#9 Calculate simple interest. 

P = float(input("Enter principle: "))
R = float(input("Enter Rate of Interest (annual): "))
T = float(input("Enter Time (annual): "))

I = (P*R*T)/100
print(I)
```
