```python
#1 Write a function that calculates Tax based on the salary 

salary = int(input("Enter your salary: "))

def tax_calc(salary):
    if salary < 0:
       return "Salary cannot be less than 0."        
    elif salary < 30_000:
        return salary*(.05)
    elif salary >= 30_000 and salary < 70_000:
        return salary*(.15)
    else:
        return salary*(.30)  

print(tax_calc(salary))    
```

```python
#2 Take two integer input print all the int numbers between them. 

num1 = int(input("Enter num1: "))
num2 = int(input("Enter num2: "))

print([n for n in range (num1, num2+1) if n%2 ==0])
```

```python
#4 return counts of digit of a number 

num = int(input("Enter number: "))

def return_count(num):
    total_digit = 0
    while num > 0:
        num = num//10
        total_digit += 1
    return total_digit

print(return_count(num))
```

```python
#5 Write a function to sum of all digits in a number 

num = int(input("Enter number: "))

def sum_digits(num):
    sum = 0
    while num > 0:
        remainder = num%10
        num//=10
        sum+=remainder
    return sum 
    
print(sum_digits(num))    
```

```python
#8 create a calculator

def calc(a, b, op):
    match op:
        case "+":
             return a+b
        case "-":
             return a-b
        case "*":
             return a*b
        case "/":
            if b == 0:
                print("Devided by 0 is not possible.")
            else:    
                return a/b
        case _: 
            print("Operation not supported!")

a = int(input("Enter number 1: "))
b = int(input("Enter number 2: "))
op = input("Enter operator: ")

calc(a, b, op)          
```

```python
#9 find Prime Number

def is_prime(n):
    for i in range(2, int(n**(1/2))):
        if n%i == 0:
            return False
        else:
            return True
            
is_prime(4)        
```
