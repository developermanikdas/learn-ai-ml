```python
#3 Create a class with some private keys And use getter and setter functions 

class Student:
    def __init__(self, name, roll_no, marks):
        # We call the setters here so validation logic isn't duplicated!
        self.set_name(name)
        self.set_roll_no(roll_no)
        self.set_marks(marks)

    # --- Name Getter & Setter ---
    def get_name(self):
        return self._name

    def set_name(self, value):
        if not value or value.strip() == "":
            raise ValueError("Name cannot be empty.")
        self._name = value

    # --- Roll Number Getter & Setter ---
    def get_roll_no(self):
        return self._roll_no

    def set_roll_no(self, value):
        if 1 <= value <= 100:
            self._roll_no = value
        else:
            raise ValueError("Roll number has to be between 1 & 100.")

    # --- Marks Getter & Setter ---
    def get_marks(self):
        return self._marks

    def set_marks(self, value):
        if value < 0:
            raise ValueError("Marks cannot be negative.")
        self._marks = value

# Testing the implementation
try:
    s1 = Student("Alice", 45, 85)
    print(f"Student: {s1.get_name()}, Roll: {s1.get_roll_no()}, Marks: {s1.get_marks()}")
    
    # This will trigger the validation
    s1.set_marks(-10) 
except ValueError as e:
    print(f"Error: {e}")
```

```python
#4 Function Overriding

import math
from abc import ABC, abstractmethod

class Shape:   
    @abstractmethod
    def area(self):
        return 0
        
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return math.pi * (self.radius**2)

        
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width

class Triangle(Shape):
    def __init__(self, height, base):
        self.height = height
        self.base = base
    def area(self):
        return .5 * self.height * self.base    
        
   
shapes = [Circle(7), Rectangle(10,5), Triangle(10,5)]

for s in shapes:
    print(f"{s.area():.2f}")
```

```python
#5 Inheritance

class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        
class Car(Vehicle):
    def __init__(self, brand, model, seats):
        super().__init__(brand, model)
        self.seats = seats 
class Bike(Vehicle):
    def __init__(self, brand, model, engine_cc):
        super().__init__(brand, model)
        self.engine_cc = engine_cc 

Car("Tata", "Nano", 4).seats        
Bike("Tata", "Nano", 1000).engine_cc        
```

```python
#6 Abstraction

from abc import ABC, abstractmethod

class Employee(ABC):
    @abstractmethod
    def calculate_salary(self):
        pass

class Intern(Employee):
    def calculate_salary(self):
        return 20_000
        
class FullTimeEmployee(Employee):
    def calculate_salary(self):
        return 40_000

class ContractEmployee(Employee):
    def calculate_salary(self):
        return 60_000

# Assigning instances to variables and printing the output
i1 = Intern()
print(f"Intern Salary: ${i1.calculate_salary():,}")

f1 = FullTimeEmployee()
print(f"Full-Time Salary: ${f1.calculate_salary():,}")

c1 = ContractEmployee()
print(f"Contract Salary: ${c1.calculate_salary():,}")
```

```python
#7 Constructor Overloading (with Default Parameters)

class Person:
    def __init__(self, name, age=None, address=None):
        self.name = name
        self.age = age
        self.address = address
        
    def display_info(self):
        info = f"Name = {self.name} "

        if self.age is not None:
            info += f"Age = {self.age}, "
        if self.address is not None:
            info += f"Address = {self.address}"    

        print(info)
        
# 1. Name only
person1 = Person("Alice")
person1.display_info()

# 2. Name + Age
person2 = Person("Bob", 30)
person2.display_info()

# 3. Name + Age + Address
person3 = Person("Charlie", 25, "123 Main St")
person3.display_info()        
```

```python
#9 Multiple inheritance

class Herbivore:
    def eat_plants(self):
        print("Foraging for berries, roots, and leaves...")

class Carnivore:
    def eat_meat(self):
        print("Hunting for fish and small mammals...")
        
class Omnivore:
    def diet_summary(self):
        print("As an omnivore, I can survive on a mixed diet!")    

class Bear(Herbivore, Carnivore, Omnivore):
    
    def __init__(self, name, species):
        self.name = name
        self.species = species
        
    def daily_routine(self):
        self.eat_plants()
        self.eat_meat()
        self.diet_summary()
        
grizzly = Bear("Barnaby", "Grizzly")

grizzly.daily_routine()
```

```python
#10 chat system

import datetime

# 1. The User Class
class User:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return self.name

# 2. The Message Class
class Message:
    def __init__(self, sender, content):
        self.sender = sender
        self.content = content
        # Automatically generate a timestamp when the message is created
        self.timestamp = datetime.datetime.now().strftime("%H:%M:%S")

    def __str__(self):
        # Format how the message will look when printed
        return f"[{self.timestamp}] {self.sender.name}: {self.content}"

# 3. The ChatRoom Class
class ChatRoom:
    def __init__(self, room_name):
        self.room_name = room_name
        self.participants = [] # List to hold User objects
        self.history = []      # List to hold Message objects

    def join(self, user):
        if user not in self.participants:
            self.participants.append(user)
            print(f"👋 {user.name} has joined '{self.room_name}'.")
        else:
            print(f"{user.name} is already in the chatroom.")

    def leave(self, user):
        if user in self.participants:
            self.participants.remove(user)
            print(f"🚪 {user.name} has left '{self.room_name}'.")
        else:
            print(f"{user.name} cannot leave because they are not in the room.")

    def send_message(self, user, content):
        # Only allow messages if the user is actually in the room
        if user in self.participants:
            new_message = Message(user, content)
            self.history.append(new_message)
        else:
            print(f"❌ Error: {user.name} must join the room before sending a message.")

    def view_history(self):
        print(f"\n--- Chat History: {self.room_name} ---")
        if not self.history:
            print("No messages yet. It's quiet in here...")
        else:
            for message in self.history:
                print(message)
        print("-" * 35 + "\n")

# --- Testing the Chat System ---

# Create a ChatRoom
tech_chat = ChatRoom("Python Dev Lounge")

# Create Users
alice = User("Alice")
bob = User("Bob")
charlie = User("Charlie")

# Users join the room
tech_chat.join(alice)
tech_chat.join(bob)

# Sending messages
tech_chat.send_message(alice, "Hey Bob, how is your OOP project going?")
tech_chat.send_message(bob, "It is going great! Just built a chatroom.")

# Charlie tries to send a message without joining
tech_chat.send_message(charlie, "Can I get some help?") 

# Charlie joins and then sends a message
tech_chat.join(charlie)
tech_chat.send_message(charlie, "Hello everyone! Happy to be here.")

# Bob leaves
tech_chat.leave(bob)

# View the final chat history
tech_chat.view_history()
```
